```python
# Mutabilidad
colors={"blanco","negro"}
colors
```

```python
# Añadir
colors.add("amarillo")
colors
```

```python
# Definimos el set 3
colors3={"amarillo"}
```

```python
colors4=colors.difference(colors3)
colors4
```

```python
# Definimos el set colors5
colors5={"blanco"}
colors4.difference_update(colors5)
colors4
```

```python
# Descartar
colors
```

```python
colors.discard("amarillo")
colors
```

```python
# Si no existe devuelve lo mismo
colors.discard("rojo")
colors
```

```python
# Si queremos eliminar algo que no existe si que avisa
colors.remove("rojo")
colors
```

```python
colors
```

```python
colors4
```

```python
# Queremos ver elementos en común
colors6=colors.intersection(colors4)
colors6
```

```python
# El orden no afecta en la intersección
colors7=colors4.intersection(colors)
colors7
```

```python
# Otra manera de hacerlo de manera más directa
colors & colors4
```

```python
colors
```

```python
# Unión
colors4.add("amarillo")
colors4
```

```python
colors.union(colors4)
```

```python
# Se puede hacer en ambas direcciones
colors4.union(colors)
```

```python
colors | colors4
```
